```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Daten laden
url = 'https://github.com/ArminVarmaz/hello-world/blob/main/Daten/wage1.xlsx?raw=true'
df = pd.read_excel(url)

# Daten für die Regression vorbereiten
X = df['educ'].values.reshape(-1, 1)  # Unabhängige Variable 'educ'
y = df['wage'].values  # Abhängige Variable 'wage'

# Lineare Regression durchführen
regression = LinearRegression()
regression.fit(X, y)

# Ergebnisse anzeigen
print('Regressionskoeffizient:', regression.coef_[0])
print('Intercept:', regression.intercept_)

# Plot der Regression
plt.scatter(X, y, color='blue', label='Datenpunkte')
plt.plot(X, regression.predict(X), color='red', label='Regression')
plt.xlabel('educ')
plt.ylabel('wage')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Daten laden
url = 'https://github.com/ArminVarmaz/hello-world/blob/main/Daten/wage1.xlsx?raw=true'
df = pd.read_excel(url)

# Daten für die Regression vorbereiten
X = df['educ']  # Unabhängige Variable 'educ'
y = df['wage']  # Abhängige Variable 'wage'

# Konstante hinzufügen
X = sm.add_constant(X)

# Lineare Regression durchführen
regression = sm.OLS(y, X)
result = regression.fit()

# Ergebnisse anzeigen
print(result.summary())

# Plot der Regression
plt.scatter(df['educ'], y, color='blue', label='Datenpunkte')
plt.plot(df['educ'], result.fittedvalues, color='red', label='Regression')
plt.xlabel('educ')
plt.ylabel('wage')
plt.legend()
plt.show()
```
